```python
import matplotlib.pyplot as plt
import pandas as pd

# Corrected DataFrame
df = pd.DataFrame({
    'item_name': ['Cappuccino', 'Cappuccino', 'Latte', 'Latte', 'Caramel Macchiato', 'Caramel Macchiato',  'Mocha', 'Mocha', 'White Mocha', 'White Mocha', 'Hot Chocolate', 'Hot Chocolate', 'Cold Coffee', 'Cold Coffee'],
    'item_size': ['Medium', 'Large', 'Medium', 'Large', 'Medium', 'Large', 'Medium', 'Large', 'Medium', 'Large', 'Medium', 'Large', 'Medium', 'Large'],
    'item_price': [3.45, 3.75, 3.45, 3.75, 4.20, 4.60, 4.00, 4.60, 4.50, 4.70, 4.20, 4.60, 3.45, 3.75]
})

# Set up colors
neon_orange = '#FF6700'
black = "#000000"
dark_grey = "#333333"

# Create the plot
fig, ax = plt.subplots(figsize=(14, 10))
fig.patch.set_facecolor(dark_grey)  # Set the canvas color

# Adjusting position of bars to avoid overlap
for idx, (size, subset) in enumerate(df.groupby('item_size')):
    bars = ax.barh(subset['item_name'] + ' (' + size + ')', subset['item_price'], color=neon_orange, alpha=0.7, label=size)

    # Add item prices on the bars
    for bar, price in zip(bars, subset['item_price']):
        ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'${price:.2f}', va='center', ha='left', color='white')

# Customize the plot
ax.set_xlabel('Item Price', color='white')
ax.set_ylabel('Item Name', color='white')
ax.set_title('Item Prices by Size', color='white')
ax.set_facecolor(black)
ax.tick_params(colors='white')
plt.xticks(color='white')
plt.yticks(color='white')

# Adding legend
plt.legend(title='Size', loc='upper right', frameon=False, prop={'size': 12, 'weight': 'bold'}, facecolor=dark_grey, edgecolor=black, labelcolor='white')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up colors
neon_orange = '#FF6700'
black = "#000000"
dark_grey = "#333333"

# Create the plot
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=df, x='item_name', y='item_price', hue='item_size')
plt.title('Item Prices by Name and Size', color='white')
plt.xticks(rotation=45, color='white')
plt.yticks(color='white')

# Customize the plot background
ax.set_facecolor(black)
plt.gcf().set_facecolor(dark_grey)
plt.xlabel('Item Name', color='white')
plt.ylabel('Item Price', color='white')
plt.legend(title='Item Size', frameon=False, prop={'size': 12, 'weight': 'bold'}, facecolor=dark_grey, edgecolor=black, labelcolor='white')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up colors
royal_blue = '#305CDE'
neon_blue = '#111184'
black = "#000000"
dark_grey = "#333333"

# Create the plot
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=df, x='item_name', y='item_price', hue='item_size', palette=[royal_blue, neon_blue])
plt.title('Item Prices by Name and Size', color='white')
plt.xticks(rotation=45, color='white')
plt.yticks(color='white')

# Customize the plot background
ax.set_facecolor(black)
plt.gcf().set_facecolor(dark_grey)
plt.xlabel('Item Name', color='white')
plt.ylabel('Item Price', color='white')
plt.legend(title='Item Size', frameon=False, prop={'size': 12, 'weight': 'bold'}, facecolor=dark_grey, edgecolor=black, labelcolor='white')

# Show the plot
plt.show()
```

```python
import numpy as np

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr = numeric_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
df['price_per_unit'] = df['item_price'] / df['item_size'].map({'Small': 1, 'Medium': 2, 'Large': 3})
```

```python
df['price_per_unit']
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price_per_unit'], kde=True, color='orange')
plt.title('Distribution of Price per Unit')
plt.xlabel('Price per Unit')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='item_name', y='price_per_unit')
plt.title('Price per Unit by Item Name')
plt.xticks(rotation=45)
plt.show()
```
